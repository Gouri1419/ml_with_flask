```python
inner_join_df=pd.merge(df1,df2,on='Customer_id',how='inner')
inner_join_df
```

```python
right_join_df=pd.merge(df1,df2,on='Customer_id',how='right')
right_join_df
```

```python
# prompt: create two dataframes int the first dataframe product and customer name in the second dataframe product name and price try to join with product name

dict1 = {'Product': ['Television', 'Television', 'Oven', 'Oven', 'Oven', 'Television'],
         'Customer Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank']}
df1 = pd.DataFrame(dict1)

dict2 = {'Product Name': ['Television', 'Oven', 'Refrigerator'],
         'Price': [1000, 500, 1200]}
df2 = pd.DataFrame(dict2)

# Merge the two dataframes on 'Product' and 'Product Name'
merged_df = pd.merge(df1, df2, left_on='Product', right_on='Product Name', how='inner')
merged_df
```
